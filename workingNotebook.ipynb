```python
import requests
from requests.adapters import HTTPAdapter
from requests.packages.urllib3.util.retry import Retry



def getArchiveMetaDataForUrl(allUrls):
    requestsClient = requests.Session()
    retries = Retry(total = 5, backoff_factor=0.5)
    requestsClient.mount('https://', HTTPAdapter(max_retries=retries))
    res = []
    counter = 0
    for url in allUrls:
        
        responseList = requestsClient.get("https://web.archive.org/cdx/search/cdx?url=" + url + "&output=json")
        for responseData in json.loads(responseList.text)[1:]:
            year = responseData[1][:4]
            month = responseData[1][4:6]
            day = responseData[1][6:8]
            responseData.append(year)
            responseData.append(month)
            responseData.append(day)
            res.append(responseData)
        counter += 1
        print(counter)
    return res
```

```python
def extractBlogLinkFromPage(page):
    res = set()
    for postDiv in page:
        links = postDiv.select("div[class^=post]")
        for link in links:
            res.add(link.find_all("a")[0]['href'])
    return res
```
