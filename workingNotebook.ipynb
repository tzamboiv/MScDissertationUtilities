```python
def makeDictFromSitemap(siteMap):
    '''Function to get a year by year picture of sitemap'''
    sitemapByYear = {i : (set(), set()) for i in range(2012,2024)}
    for i in siteMap[1:]:
        sitemapByYear[int(i[0])][0].add(i[1])
        sitemapByYear[int(i[0])][1].add(i[2])
    return sitemapByYear
```

```python
def unionAllBlogUrls(siteMap):
    '''Creates set with all URLs from SiteMap'''
    res = set()
    siteMapDict = makeDictFromSitemap(siteMap)
    for year in siteMapDict.keys():
        res = res | siteMapDict[year][1]
    return res
```

```python
def cleanAndFormUrl(waybackMachineUrlKey, rootUrl):
    '''Cleans urlKey and formats as url'''
    if rootUrl == "blog.predpol.com":
        #removes the urlKey and replaces with the root of the URL
       url = rootUrl + waybackMachineUrlKey.split('com,predpol,blog)')[1]
    elif rootUrl == "predpol.com":
        #removes the urlKey and replaces with the root of the URL
        url = rootUrl + waybackMachineUrlKey.split('com,predpol)')[1]
    else:
        #Throws error is invalid rootUrl is used
        raise ValueError("rootUrl only excepts 'predpol.com' or 'blog.predpol.com'")
    #Logic in return statement removes any Url parameters (tracking strings and things of that nature)
    return url.split("?")[0]

#Create a request client to customize retry and backoff for API
requestsClient = requests.Session()
retries = Retry(total = 5, backoff_factor=0.5)
requestsClient.mount('https://', HTTPAdapter(max_retries=retries))

#responseList = requestsClient.get("https://web.archive.org/cdx/search/cdx?url=" + url + "&output=json")

def getEarliestTimeStampOfArchivedBlog(url, requestClient = requestsClient):
    '''Queries web.archive.org cdx API to find the earliest timestamp for the page in Wayback Machine'''
    requestUrl = "https://web.archive.org/cdx/search/cdx?url=" + url + "&output=json&limit=1"
    print(requestUrl)
    response = requestsClient.get(requestUrl)
    return json.loads(response.text)[1][1]

def categorizeBlogUrl(url):
    '''I'm only interested in blog posts, so I want to label these'''
    if "/page/" in url:
        return 'pageList'
    elif '/tag/' in url:
        return 'tag'
    elif url == 'blog.predpol.com':
        return 'root'
    else:
        return 'blog'

def makeUrlKeyDataFrame(urlKeySet, rootUrl):
    '''NOTE: This is not a pure function. It changes and is changed by the resultList and seenUrls variables. 
    These variables allow this function to gracefully(ish) crash if the API requests falls and restart'''

    uniqueUrlSet = set()
    #clean the URLs
    for urlKey in urlKeySet:
        url = cleanAndFormUrl(urlKey, rootUrl)
        uniqueUrlSet.add(url)
    totalSize = len(uniqueUrlSet)
    print(totalSize)
    counter = 1
    for url in uniqueUrlSet:
        if url in seenUrls:
            #This allows the function to skip if any data is already collected
            print('skipping')
            counter += 1
            continue
        timestamp = getEarliestTimeStampOfArchivedBlog(url)
        pageType = categorizeBlogUrl(url)
        resultList.append([url, timestamp, pageType, timestamp[:4], timestamp[4:6], timestamp[6:8]])
        seenUrls.append(url)
        print(str(counter) + '/' + str(totalSize))
        counter += 1
    df = pd.DataFrame(resultList)
    df.columns = ['url','timestamp','pageType', 'year', 'month', 'day']
    return df
```

```python
def getBlog(url, timestamp, requestClient = requestsClient):
    '''Gets the given blog at the given timestamp'''
    requestUrl = 'https://web.archive.org/web/' +str(timestamp) + '/' + str(url)
    print(requestUrl)
    response = requestsClient.get(requestUrl)
    print(response.status_code)
    return response

def writeBlogToFile(url, blogString):
    '''Writes blog to file with name of blog'''
    with open(str(url).replace("/", "_") + '.html', "w+") as text_file:
        text_file.write(blogString)
    return 'Success'

def scrapeAllBlogs(df):
    '''Scrapes all blogs from the dataframe'''
    #Only scrape URLs categorized as 'blog'
    onlyBlogs = df[df.pageType == 'blog']
    blogUrls = list(onlyBlogs.url)
    blogTimestamps = list(onlyBlogs.timestamp)
    print(len(blogUrls))
    print(len(blogTimestamps))
    #Runs while loop that breaks when the blogURL list is empty list
    while len(blogUrls) != 0:
        url = blogUrls.pop()
        print(url)
        timestamp = blogTimestamps.pop()
        print(timestamp)
        try:
            #If these succeed, then blogUrls and blogTimestamps are successfully deincremented
            blogString = getBlog(url, timestamp)
            writeBlogToFile(url, blogString.text)
        except Exception as e: 
            #If there is an exception, the url and timestamp are readded to the list
            print(e)
            blogUrls.append(url)
            blogTimestamps.append(timestamp)
        finally:
            #Prints the list as a tracking mechanism when code is run
            print(len(blogUrls))
            print(len(blogTimestamps))
```

```python
def extractBlogTestFromBlogDotPredPolDotCom(htmlString):
    '''Takes HTML from a blog, then converts it to a BeautifulSoup Object
    Extracts a title from the post-header div and then extracts blog body from post-body divs'''
    soup = BeautifulSoup(htmlString)
    #Extract Title of Blog
    blogTextRes = [soup.find_all("div", {"class": "section post-header"})[0].find_all('h1')[0].text + ':']
    
    
    #Extract Text from Blog Body
    blogBodyTextTags = soup.find_all("div", {"class": "section post-body"})[0].find_all('span')[0].find_all(['p', 'h2'])
    for tag in blogBodyTextTags:
        line = tag.text
        if line == 'Sources':
            break
        if tag.name == 'h2':
            line = line + ':'
        blogTextRes.append(line)
    blogText = ' '.join(blogTextRes)
    blogText = blogText.replace(r"  ", ' ')
    blogText = unidecode.unidecode(blogText)
    return blogText

def readBlogFromFile(file):
    '''Reads blog file and returns string'''
    with open(file, 'r') as fileText:
        res = fileText.read()
        return str(res)

def extractAllBlogs(absoluteDirectory, extractionFunction):
    '''Takes an absolute file directory, extracts text from all blogs saved there
    Returns a tuple with a list of blogs strings and a list of file names that threw errors
    Pass the extraction function that is appropriate for the type of blog'''
    fileList = os.listdir()
    res = []
    failed = []
    for file in fileList:
        if file == '.DS_Store':
            continue
        try:
            blogHTML = readBlogFromFile(file)
            blogText = extractionFunction(blogHTML)
            res.append(blogText)
        except Exception as e:
            print(e)
            failed.append(file)
    return (res, failed)

def writeBlogStringToTxtFile(blogString, filename):
    with open(filename, "w+") as text_file:
        text_file.write(blogString)
    return
    
```

```python
def stripUrl(url):
    '''Remove :80 port specification from url'''
    #copyUrl = url.split("?")[0]
    urlSplitOnPort = url.split(":80")
    copyUrl = ''.join(urlSplitOnPort)
    return copyUrl

def categorizeUrl(url):
    '''Very manual url categorizer IN PROGRESS'''
    cat = ""
    url = stripUrl(url)
    #Regex to match Blog Table of Contents Pages
    blogMonthYearPagesRegex = re.compile('/[0-9]{4}/[0-9]{2}')
    #Regex to Match blog ID pages
    blogIdPage = re.compile('\?p=[0-9]+')
    #Search Regex
    searchRegex = re.compile('\?s=')
    #Wordpress Asset
    wordpressRegex = re.compile('wp-')
    #Tag page
    tagPageRegex = re.compile('/tag/')
    #Category page
    categoryPageRegex = re.compile('/category')
    #about page
    aboutRegex = re.compile('/about/')

    if re.search(blogMonthYearPagesRegex, url):
        cat = "2012-2014BlogTableOfContents"
    elif re.search(blogIdPage, url):
        cat = 'blogId'
    elif re.search(searchRegex, url):
        cat = 'search'
    elif re.search(wordpressRegex, url):
        cat = 'wordpressAsset'
    elif re.search(tagPageRegex, url):
        cat = 'tag'
    elif re.search(categoryPageRegex, url):
        cat = 'category'
    elif re.search(aboutRegex, url):
        cat = 'about'
    return cat

            
```
